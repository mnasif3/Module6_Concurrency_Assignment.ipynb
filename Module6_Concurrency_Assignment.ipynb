### 13.1
This task writes the current date as a string to a text file named today.txt.

```python
from datetime import date

today = date.today()

with open("today.txt", "w") as file:
    file.write(str(today))

print("Date written to today.txt")
```

### 13.2
This task reads the contents of today.txt into a string variable called today_string.

```python
with open("today.txt", "r") as file:
    today_string = file.read()

print("today_string:", today_string)
```

### 13.3
This task parses the date string into a datetime object.

```python
from datetime import datetime

parsed_date = datetime.strptime(today_string, "%Y-%m-%d")

print("Parsed date:", parsed_date)
```

### 15.1
This task uses multiprocessing to create three separate processes. Each process waits a random time between 0 and 1 second, prints the current time, and exits.

```python
from multiprocessing import Process
import time
import random
from datetime import datetime

def worker(process_id):
    """Function that each process will run"""
    # Wait a random time between 0 and 1 second
    wait_time = random.uniform(0, 1)
    time.sleep(wait_time)
    
    # Print current time
    current_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S.%f")[:-3]
    print(f"Process {process_id}: Waited {wait_time:.3f} seconds, Current time: {current_time}")

# Create three separate processes
if __name__ == '__main__':
    processes = []
    
    for i in range(3):
        p = Process(target=worker, args=(i+1,))
        processes.append(p)
        p.start()
    
    # Wait for all processes to finish
    for p in processes:
        p.join()
    
    print("All processes completed!")
```
